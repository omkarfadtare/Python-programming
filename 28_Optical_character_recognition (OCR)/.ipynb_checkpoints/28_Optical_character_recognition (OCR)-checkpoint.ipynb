## Optical Character Recognition (OCR)
- Optical character recognition is the process of reading text from images into machine readable format.
- OCR as a process generally consists of several sub-processes:
  - __`Preprocessing of the Image`__
  - __`Text Localization`__
  - __`Character Segmentation`__
  - __`Character Recognition`__
  - __`Post Processing`__
- OCR engines: Tesseract OCR, OCRopus, Ocular, SWiftOCR
- Cloud based tools for OCR: Google Cloud Vision,  Amazon Textract,  IBM Watson OCR, etc.
- On-premises tools for OCR: Pytesseract, Capture2Text, etc.
- Its main aim to identify and capture all the unique words by leveraging the combination of deep models and huge datasets publicly available, deep models achieve state-of-the-art accuracies on given tasks.
- Usecases of OCR:
  - Building license plate readers
  - Digitizing invoices
  - Digitizing menus
  - Digitizing ID cards

- You can use different psm (Page Segmentation Mode) and oem (OCR engine mode). 
- You can also choose the language.

__Page Segmentation Modes:__
- 0: Orientation and script detection (OSD) only.
- 1: Automatic page segmentation with OSD.
- 2: Automatic page segmentation, but no OSD or OCR.
- 3: Fully automatic page segmentation, but no OSD. (default)
- 4: Assume a single column of text of variable sizes.
- 5: Assume a single uniform block of vertically aligned text.
- 6: Assume a single uniform block of text.
- 7: Treat the image as a single text line.
- 8: Treat the image as a single word.
- 9: Treat the image as a single word in a circle.
- 10: Treat the image as a single character.
- 11: Sparse text. Find as much text as possible in no particular order.
- 12: Sparse text with OSD.
- 13: Raw line. Treat the image as a single text line, bypassing hacks that are Tesseract-specific.
- 14: Raw line with OSD.
- 15: Sparse text. Sparse text with OCR.
- 16: Sparse text. Raw line with OCR.

__OCR Engine Modes:__
- 0: Original Tesseract only.
- 1: LSTM only.
- 2: Combines both the original Tesseract and LSTM.

```python
# Example2: Advanced way

import cv2
import pytesseract
import numpy as np
from PIL import Image

path = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # tesseract engine path
pytesseract.pytesseract.tesseract_cmd = path

# PReprocessing image
def preprocess_image(image_path):
    image = cv2.imread(image_path)

    # Converting an image from color image to grayscale image (In OpenCV, the default color order is BGR : In Pillow, the default color order is RGB)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Applying thresholding to create a binary image
    _, binary_image = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    return binary_image

def extract_text_from_invoice(image_path):
    preprocessed_image = preprocess_image(image_path)

    extracted_text = pytesseract.image_to_string(preprocessed_image, lang='eng', config='--psm 6')
    return extracted_text

invoice_path = r'E:\Github\Python\28_Optical_character_recognition (OCR)\Images\5.png'
result = extract_text_from_invoice(invoice_path)

print(result)
```

```python
# Example3: Advanced way

import cv2
import pytesseract
import numpy as np
from PIL import Image

path = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # tesseract engine path
pytesseract.pytesseract.tesseract_cmd = path

# Preprocessing for Tesseract: involves rescaling, binarization, noise removal, deskewing, etc.
image_path = r'E:\Github\Python\28_Optical_character_recognition (OCR)\Images\1.png'
image = cv2.imread(image_path)

# Preprocessing for Tesseract: involves rescaling, binarization, noise removal, deskewing, etc.

# This function converts the input color image to grayscale image
def get_grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# This function reduces noise in an image by applying median blur technique: Median blur is a nonlinear image processing technique 
# that is particularly effective at reducing randomly occurring bright and dark pixels in an image
def remove_noise(image):
    return cv2.medianBlur(image,5)
 
# This function applies thresholding to the image: Thresholding is a process where we convert an image into a binary image,
# meaning every pixel is assigned either a value of 0 (black) or 255 (white), based on whether its intensity is below or above 
# a certain threshold.
# Otsu's method is a way to automatically find an optimal threshold value for the conversion. Otsu's method helps make the text 
# stand out more clearly from the background in an image.
def thresholding(image):
    return cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# This function performs dilation on the image: Dilation can be used to make text characters bolder and more connected. 
# Imagine you have a white letter "A" on a black background. If you apply dilation, the letter "A" becomes thicker because
# the bright pixels at the edges of the letter expand into the surrounding black background.
def dilate(image):
    kernel = np.ones((5,5),np.uint8)
    return cv2.dilate(image, kernel, iterations = 1)
    
# This function performs erosion on the image: Erosion can be used to make text characters thinner and separate them from each other.
# Imagine you have a thick letter "A" on a black background. If you apply erosion, the edges of the letter "A" become thinner 
# because the bright pixels at the edges are gradually removed.
def erode(image):
    kernel = np.ones((5,5),np.uint8)
    return cv2.erode(image, kernel, iterations = 1)

# This function applies an opening operation to the image (erosion followed by dilation): Opening is a combination of two basic 
# operations erosion followed by dilation.
# Imagine you have a picture with tiny specks of dust (noise) and small dots. If you apply an opening operation, the dust and
# small dots might disappear, leaving behind the larger and more meaningful parts of the image.
def opening(image):
    kernel = np.ones((5,5),np.uint8)
    return cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# This function applies the canny edge detection algorithm to the image: Canny edge detection is an image processing technique 
# that helps identify the edges or boundaries of objects within an image.
# Imagine you have a picture of a cat. If you apply Canny edge detection, the algorithm will highlight the edges of the cat, 
# making it stand out from the background.
def canny(image):
    return cv2.Canny(image, 100, 200)

# This function performs deskewing on the image: Deskewing is the process of straightening or aligning skewed (tilted) text in an image. 
# Imagine you have a scanned document with slightly tilted paragraphs of text. If you apply deskewing, the algorithm will automatically
# rotate the image to make the text lines perfectly horizontal
def deskew(image):
    coords = np.column_stack(np.where(image > 0))
    angle = cv2.minAreaRect(coords)[-1]
    if angle < -45:
        angle = -(90 + angle)
    else:
        angle = -angle
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)
    return rotated

# This function performs template matching: template matching might be used to locate specific symbols or characters in an image,
# especially when the appearance of those symbols is known beforehand.
# Imagine you have a picture of a red apple, and you have a small template of a red apple. If you apply template matching, 
# the algorithm will highlight areas in the larger image where the template (small red apple) is found.
def match_template(image, template):
    return cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED) 
```

```python
# Apply preprocessing functions
gray = get_grayscale(image)
noise = remove_noise(gray)
thresh = thresholding(noise)
dilate_image = dilate(thresh)
erode_image = erode(dilate_image)
opened_image = opening(erode_image)
canned_image = canny(opened_image)
deskewed = deskew(canned_image)

# Save the deskewed image
deskewed_path = 'deskewed_image.png'
cv2.imwrite(deskewed_path, deskewed)

# Apply template matching
template_result = match_template(deskewed, deskewed)

# Save the template matching result
template_result_path = 'template_matching_result.png'
cv2.imwrite(template_result_path, template_result)

# Apply Tesseract OCR to the saved deskewed image
custom_config = r'--oem 3 --psm 6'
result = pytesseract.image_to_string(deskewed_path, config=custom_config)
print(result)
```

```python
# Selecting recognised characters with confidence greater than 50

import cv2
import pytesseract
from pytesseract import Output

path = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # tesseract engine path
pytesseract.pytesseract.tesseract_cmd = path

image_path = r'E:\Github\Python\28_Optical_character_recognition (OCR)\Images\7.jpg'  # image path
img = cv2.imread(image_path)

custom_config = r'-l eng --psm 6 --oem 3'  # custom configuration

data = pytesseract.image_to_data(img, config=custom_config, output_type=Output.DICT)  # applying boxes around recognized words

# Extract characters with confidence greater than 80
selected_characters = ""
for i in range(len(data['text'])):
    text = data['text'][i]
    conf = data['conf'][i]
    if isinstance(conf, str):
        confidences = conf.split()
        for char, conf in zip(text, confidences):
            if float(conf) > 50:
                selected_characters += char
    else:
        if float(conf) > 80:
            selected_characters += text

print(selected_characters)
```

```python
# Detecting orientation and script of an image

import cv2
import pytesseract
import re

path = r'C:\Program Files\Tesseract-OCR\tesseract.exe' # tesseract engine path
pytesseract.pytesseract.tesseract_cmd = path

img = cv2.imread(r'E:\Github\Python\28_Optical_character_recognition (OCR)\Images\8.jpg')

osd = pytesseract.image_to_osd(img, config="--dpi 300")
angle = re.search('(?<=Rotate: )\d+', osd).group(0)
script_match = re.search('(?<=Script: )\d+', osd)
print("Angle:", angle)
if script_match:
    script = script_match.group(0)
    print("script:", script)
else:
    print("Script information not found.")
```
